# New-Wheels Business Report Visualizations

```python
# import libraries for data manipulation
import numpy as np
import pandas as pd

# import libraries for data visualization
import matplotlib.pyplot as plt
import seaborn as sns
```

### Question 1:

```python
df = pd.read_csv('C://Users/15612/Downloads/Week 1 SQL/Views/New_Wheels_Project_Question_1.csv')
```

```python
df.head(5).plot.bar(x='STATE', y='CUSTOMER_COUNT', rot=0, color='black')
plt.title("Top 5 States with Most Customers")
plt.xlabel("State")
plt.ylabel("Customers")
plt.xticks(rotation=30, horizontalalignment="center");
```
